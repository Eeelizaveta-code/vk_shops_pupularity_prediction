## Подсчет метрик

**Конверсия clicks/views по категориям**

```python
df_stats_merged = df_stats.merge(df_placements[['hash_placement_id', 'hash_offer_id']],
                                 on='hash_placement_id',
                                 how='left')

df_stats_merged = df_stats_merged.merge(df_offers[['hash_offer_id', 'category']],
                                        on='hash_offer_id',
                                        how='left')

category_ctr = (
    df_stats_merged.groupby('category')
    .agg({
        'views': 'sum',
        'clicks': 'sum'
    })
    .assign(clicks_to_views=lambda x: x['clicks'] / x['views'])
    .sort_values('clicks_to_views', ascending=False)
)

category_ctr
```

**Конверсия orders/clicks по категориям**

```python
df_orders_clean = df_orders[df_orders['order_status_code']=='5']

df_orders_merged = df_orders_clean.merge(
    df_placements[['hash_placement_id', 'hash_offer_id']],
    on='hash_placement_id',
    how='left'
)

df_orders_merged = df_orders_merged.merge(
    df_offers[['hash_offer_id', 'category']],
    on='hash_offer_id',
    how='left'
)

orders_by_category = (
    df_orders_merged
    .groupby('category')
    .agg(orders=('hash_order_id', 'count'))
    .reset_index()
)

clicks_by_category = (
    df_stats_merged
    .groupby('category')
    .agg(clicks=('clicks', 'sum'))
    .reset_index()
)

cr2_df = orders_by_category.merge(clicks_by_category, on='category', how='left')

cr2_df['orders_to_clicks'] = cr2_df['orders'] / cr2_df['clicks'].replace(0, np.nan)
cr2_df['orders_to_clicks'] = cr2_df['orders_to_clicks'].fillna(0)

cr2_df = cr2_df.sort_values('orders_to_clicks', ascending=False).reset_index(drop=True)

cr2_df
```

Совместная конверсия: orders/views по категориям

```python
full_conv = category_ctr.reset_index().merge(
    cr2_df[['category', 'orders_to_clicks']],
    on='category',
    how='left'
)

full_conv['orders_to_views'] = (
    full_conv['clicks_to_views'] * full_conv['orders_to_clicks']
)

full_conv = full_conv.sort_values('orders_to_views', ascending=False).reset_index(drop=True)

full_conv
```

И еще после этого всего хотелось бы посмотреть  **views в процентах от общего числа** .

```python
total_views = category_ctr['views'].sum()
views_percent_clean = (
    category_ctr[['views']]
    .assign(views_pct=lambda x: x['views'] / total_views * 100)
    .sort_values('views_pct', ascending=False)
    .reset_index()
)

views_percent_clean

```

```python
placements_with_cat = df_placements.merge(
    df_offers[['hash_offer_id', 'category']],
    on='hash_offer_id',
    how='left'
)

placements_by_category = (
    placements_with_cat
    .groupby('category')
    .agg(placements_count=('hash_placement_id', 'count'))
    .reset_index()
)

total_placements = placements_by_category['placements_count'].sum()

placements_by_category['placements_pct'] = (
    placements_by_category['placements_count'] / total_placements * 100
)

placements_by_category = (
    placements_by_category
    .sort_values('placements_count', ascending=False)
    .reset_index(drop=True)
)

placements_by_category
```

```python
views_by_category = (
    category_ctr[['views']]
    .reset_index()
)

placements_table = placements_by_category[['category', 'placements_count']]

views_vs_placements = views_by_category.merge(
    placements_table,
    on='category',
    how='left'
)

views_vs_placements['views_per_placement'] = (
    views_vs_placements['views'] / views_vs_placements['placements_count']
)

views_vs_placements = (
    views_vs_placements
    .sort_values('views_per_placement', ascending=False)
    .reset_index(drop=True)
)

views_vs_placements
```

```python
clicks_by_category = (
    category_ctr[['clicks']]
    .reset_index()
)

placements_table = placements_by_category[['category', 'placements_count']]

clicks_vs_placements = clicks_by_category.merge(
    placements_table,
    on='category',
    how='left'
)

clicks_vs_placements['clicks_per_placement'] = (
    clicks_vs_placements['clicks'] / clicks_vs_placements['placements_count']
)

clicks_vs_placements = (
    clicks_vs_placements
    .sort_values('clicks_per_placement', ascending=False)
    .reset_index(drop=True)
)

clicks_vs_placements
```

```python
orders_table = orders_by_category.copy()

placements_table = placements_by_category[['category', 'placements_count']]

orders_vs_placements = orders_table.merge(
    placements_table,
    on='category',
    how='left'
)

orders_vs_placements['orders_per_placement'] = (
    orders_vs_placements['orders'] / orders_vs_placements['placements_count']
)

orders_vs_placements = (
    orders_vs_placements
    .sort_values('orders_per_placement', ascending=False)
    .reset_index(drop=True)
)

orders_vs_placements
```

По факту «Хобби и развлечения» генерируют большой интерес на платформе, но значительно хуже конвертируют этот интерес в реальные покупки — именно поэтому категория проседает до 7-го места в порядке *orders_to_views*. *  *Спорт и отдых.* Категория занимает третье место по количеству контента. Второе место по количеству просмотров на единицу контента - отличный показатель. Но проседает на конверсии clicks_to_views, занимая там девятое место, и в orders_to_clicks, падая на 12ое. По итогу, стоя на #11 по числу заказов на одно размещение, эта категория не приносит большой выручки при значительных затратах на размещение. Это своего рода "серая зона". То есть видимо люди видят товар в ленте, но у них не возникает мнгновенного испульса его купить. Однако многие авторы любят публиковать ролики с этой категорией. Поэтому здесь лучше сконцентрироваться на вопросах создания ролика/поста, как сдлать его более цепляющим. Поэтому мы бы предлоржили снизить процент контента по этой теме, но улучшить его качество - это окажет более благотворное влияние на аудиторию.

*  *Бытовая техника*. Все показатели ниже среднего. Размещений на платформе много (#5), но эффективность низкая. Views_to_placement низкий – пользователи не выражают интереса к товарам. Из-за этого и все послежующие показатели в цепочке низкие. В целом категория плохо монетизируется в s-commerce, и потому потенциала к масштабированию мы не наблюдаем. Предположительно, это снова "серая зона", которую авторы еще не научились продавать в постах. Потому есть смысл постепенно сокращать объем размещений, делая фокус на качестве и форматах, а не на количестве. Оставить только точечные сильные интеграции.

* *Детские товары*. Те же выводы, что и про бытовую технику. Количественно уменьшаем, экспериментируем с форматами - "серая зона".

* *Канцелярские товары* и *Транспорт*. Оба изначально занимают небольшие проценты рынка, 2,3 и 2,9 соотв. По показателям чуть ниже среднего. Потенциала к масштабированию не проявляют. У обоих узкая и четко очерченная целевая аудитория. Не рекомендуем к масштабированию, однако предлагаем сконцентрироваться на умных алгоритмах и персональных рекомендациях. Для канцелярских товаров - студенты и школьники, и авторы, популярные среди этой ЦА. Для транспорта - автомобилисты.

2. **Товары для животных** \
Категория показывает сильные показатели и в просмотрах, и в заказах, при этом недонасыщена контентом, поэтому её нужно масштабировать постепенным увеличением количества размещений.
3. **Дом и дача** \
Несмотря на высокий объём контента, категория остаётся одной из самых успешных по выручке, но требует улучшения качества постов для увеличения кликов. К масштабированию рекомендуется.
4. **Продукты питания** \
Категория прибыльная и недозагружена, поэтому её стоит постепенно масштабировать, чтобы усилить влияние на заказы.
5. **Хобби и развлечения** \
Категория генерирует высокий интерес (много просмотров и кликов), но плохо конвертирует клики в заказы, поэтому нужно улучшать качество постов, а не увеличивать объём.
6. **Канцелярские товары** \
Категория имеет показатели чуть ниже среднего и узкую ЦА, поэтому не подходит для масштабирования, но нуждается в персонализированных подборках для школьников и студентов.
7. **Транспорт** \
Как и канцелярия, категория узконаправленная, со средними метриками, поэтому масштабировать её нецелесообразно, но можно улучшить точность рекомендаций автоаудитории.
8. **Детские товары** \
Эта категория относится к «серой зоне»: метрики ниже среднего, поэтому объём снижать, а контент — улучшать и тестировать форматы.
9. **Гардероб** \
Категория перенасыщена контентом: много размещений, много просмотров, но низкий views_per_placement, поэтому увеличивать объём контента нельзя — нужно снижать.
10. **Бытовая техника** \
Категория демонстрирует слабый интерес аудитории и низкую монетизацию, поэтому объём размещений нужно сокращать и оставлять только точечные качественные интеграции.
11. **Спорт и отдых** \
Категория популярна у авторов и по просмотрам, но плохо конвертирует в заказы, поэтому стоит уменьшить количество размещений и работать над более цепляющим контентом.
12. **Товары для геймеров** \
Аудитория слабо представлена в VK (она находится на других платформах), просмотры низкие, поэтому категорию можно исключить из каталога и заменить рекламой на VK Play.
13. **Все товары** \
Категория не является самостоятельной товарной группой, и показатели указывают, что она не имеет потенциала к масштабированию — её стоит оставить без изменений.
14. **Ремонт и строительство** \
Категория имеет высокое количество кликов, но очень слабую конверсию в заказы, поэтому она не подходит для масштабирования и требует улучшения продающего контента.
15. **Электроника** \
Электроника получает много кликов, но почти не конвертируется в покупки, из-за чего категория экономически неэффективна и занимает слишком большой объём — её нужно сокращать.

# *Как мотивировать блогеров выбирать конкретные товары?*


## ЧАСТЬ 1: МОДЕЛИ ОПЛАТЫ И ФИНАНСОВЫЕ СТИМУЛЫ

### 1. Performance-Based Payment (Оплата за результат)

Суть: Оплата зависит от конкретных метрик — клики, заказы, или продажи.

Применение в VK social commerce:

- Модель CPO (Cost Per Order): Продавец платит только за реальные заказы
- Модель CPA (Cost Per Action): Оплата за клик, подписку, добавление в корзину
- Комиссия от продажи: Блогер получает процент от каждой совершённой через его ссылку покупки

Примеры из практики:

- VK AdBlogger через Ozon: Блогеры получают от 8 до 60 000 баллов за публикацию ВКонтакте (зависит от размера аудитории)
- YouTube Shopping: Комиссии составляют 5-20% от стоимости товара, с возможностью увеличения до 15% для топ-креаторов с высокими конверсиями
- Amazon Affiliate: Комиссии варьируются от 1% до 20% в зависимости от категории (fashion — 10%, beauty — 5-10%, электроника — 1-2%)
- Xiaohongshu (RedNote): На платформе используются комиссионные структуры 10-15% для подтверждённых продаж

Почему это работает:

- Прямая связь между действиями блогера и его заработком
- Мотивирует выбирать товары, которые реально конвертируют
- Справедливо — блогер получает больше, если его контент лучше
- Минимизирует риск для продавца

Недостатки:

- Блогеры могут отказаться — слишком высокий риск, если они сомневаются в товаре
- Требует прозрачного отслеживания (UTM, промокоды, реферальные ссылки)

Рекомендация для VK:

Базовая комиссия: 8-12% от стоимости товара
Премиум комиссия: 15-18% для топ-креаторов с высокой конверсией
Бонусы за количество: +2% за 10+ продаж в неделю

---

### 2. Hybrid Payment Model (Гибридная модель)

Суть: Комбинация фиксированной оплаты и бонусов за производительность.

Структура:

Фиксированная база: 5,000-50,000 рублей в месяц (зависит от размера аудитории)
Плюс комиссия: 8-10% от каждой продажи
Плюс бонусы за достижение целей: +5,000 рублей за 50 заказов в месяц

Почему это лучше:

- Блогер получает стабильный доход (не боится рисков)
- Мотивирован работать лучше (есть бонусы)
- Бренд знает минимальные затраты (фиксированная часть)
- Данные показывают: гибридные модели в 2025 году признаны самыми эффективными

Пример реализации:

- Douyin/Xiaohongshu: Топ-креаторы получают базовую ставку за рекламное сообщение плюс процент от GMV
- TikTok Shop: Программы типа "GMV Sprint" предлагают бонусы в размере $500 за достижение определённых целей плюс основной заработок

Рекомендация для VK:

Категория "Дом и дача" (большой объём, высокие заказы):
  Мини-блогер (1-10K): 3,000₽ в месяц плюс 10% комиссия
  Микро-блогер (10-100K): 8,000₽ в месяц плюс 10% комиссия

### 1. Intelligent Matching System (Умное сопоставление)

Суть: Алгоритм предлагает блогеру товары, которые подходят ему и его аудитории.

Как работает:

INPUT (Входные данные):
  История контента блогера (последние 50 постов)
  Демография его аудитории
  Категории, которые он часто рекомендует
  Его конверсионные метрики по каждой категории

ALGORITHM (Алгоритм):
  Анализирует, по каким категориям блогер показывает 5% и выше конверсию
  Выбирает новые товары из этих категорий
  Проверяет, что товар релевантен его аудитории
  Ранжирует по вероятности успеха

OUTPUT (Вывод):
  Блогеру в кабинете:
  "Топ-3 товара для вас эту неделю:"
    1. Товар A (красота, ваша конверсия 8%) — комиссия 12%
    2. Товар B (животные, вероятность успеха 85%) — комиссия 15%
    3. Товар C (дом, похож на товар, который продал 50 шт.) — комиссия 10%

Преимущества:

- Блогеру не нужно искать товары — они приходят к нему
- Блогер получает товары, которые ему работают
- Платформа видит, что готовит товар с высокой вероятностью продажи
- Мотивация: "Смотри, вот товары, которые подходят твоей аудитории"

## Сегментирование аудитории

**Сегментирование аудитории и bottlenecks**
- Посты с видео и интерактивным контентом имеют высокий охват и вовлечение.
- Категории с наибольшей конверсией (orders_to_clicks) — "Красота и здоровье", "Товары для животных", "Продукты питания", "Дом и дача".
- CTR (отношение кликов к просмотрам) выше всего для категорий "Хобби и развлечения", "Канцелярские товары", "Ремонт и строительство".
- Наиболее выраженные bottlenecks наблюдаются в переходах от просмотров к кликам, а также к заказам, где конверсия значительно снижается.
- К примеру "Электроника" и "Гардероб" имеют очень большие просмотры, но конверсия меньше, чем у "Красоты и здоровья".
- Для категории "Красота и здоровье" и "Товары для животных" — эффективны видео-обзоры, распаковки и честные отзывы.
- Для категорий "Гардероб", "Дом и дача" — полезны посты с тематическими челленджами, лайфхаками и DIY (сделай сам).
- Активно работают форматы с геймификацией: значки, лидерборды, челленджи, а также эксклюзивный доступ к продуктам через сообщества и VIP-статусы.




*Сегменты (segment_code): Значения от 1 до 6. Судя по контексту, это, скорее всего, уже предварительно вычисленные сегменты, где 1 — наименее активные, а 6 — наиболее активные пользователи. Mean = 3.23 говорит о том, что средний пользователь находится в сегменте 3.*


Мы будем использовать комбинированную сегментацию:
1.  Уровень общей активности (на основе segment_code и общего вклада).
2.  Тип создаваемого контента (Создатели постов vs Создатели клипов vs Универсалы).

**Давайте выделим 6 ключевых сегментов:**

*Сегментировав сначала по segment_code, а потом верхнюю и нижнюю половину сегмента по популярности*

1. Ядро платформы (Top Creators)
- segment_code = 5-6. Самые активные и влиятельные пользователи.
- Портрет: Генерируют непропорционально большой объем просмотров и вовлеченности. Часто блогеры, эксперты, медийные личности.
- Цель: Удержать, мотивировать на создание еще более качественного контента, превратить в партнеров.
- Способы поощрения: персональный менеджер, ранний доступ к фичам, монетизация/рекламные контракты, помощь в продвижение через алгоритмизацию, углубленный анализ активности аудитории и трендах, верификация (галочка)


2. Активные энтузиасты (Active Contributors)
- segment_code = 3-4. Регулярные пользователи, основа комьюнити.
- Портрет: Стабильно публикуют контент (в основном посты), имеют стабильную, но не огромную аудиторию. Комментируют и лайкают.
- Цель: Стимулировать рост, помочь им перейти в сегмент "Ядра", поощрять эксперименты с форматами (например, с клипами).
- Способы поощрения: углубленный анализ активности аудитории и трендах, возможность добавлять в посты больше медиа и создавать опросы, вебинары и гайды от топовых создателей о контентном росте, система баджей/achievements, приоритет в очередях на модерацию

3.  Массовая аудитория (Casual Users)
- Кто: segment_code = 1-2. Наиболее многочисленный сегмент.
- Портрет: Потребляют больше, чем создают. Публикуют мало (медиана 76 постов в 30 дней — это ~2-3 поста в день, что немного в контексте общего объема). Почти не публикуют клипы.
- Цель: Активировать, обучить, подтолкнуть к созданию первого контента, особенно в новом формате (клипы).
- Способы поощрения: онбординг, гайды "Как создать первый пост/клип", "Как набрать первых подписчиков", шаблоны для клипов, напоминания, уведомления о трендах, которые могут их заинтересовать, программа поощрений за первое действие


4.  Создатели клипов (Clip Specialists)
- Кто: Пользователи из любых сегментов (но в основном 3-6), у которых `clips_30d` > 75-й перцентиль (т.е. > 4 клипов в месяц), но при этом их активность в постах может быть средней или низкой.
- Портрет: Сфокусированы на самом "горячем" формате, который дает высокий охват.
- Цель: Поощрять и упрощать создание именно этого формата, так как он драйвит рост платформы.
- Способы поощрения: приоритетный показ их новых клипов в рекомендательной ленте, расширенная библиотека музыки, стикеров, фильтров без ватермарок, приглашение к участию в платных спонсируемых челленджах.


5.  Универсальные создатели (Hybrid Creators)
- Кто: Пользователи из сегментов 3-6, которые стабильно публикуют и посты, и клипы (`posts_30d` и `clips_30d` выше медианы).
- Портрет: Наиболее ценны для платформы, так как диверсифицируют контент и удерживают разную аудиторию.
- Цель: Выделять как образец для подражания, поощрять максимально.
- Способы поощрения: Комбинация лучших "плюшек" из сегментов "Активные энтузиасты" и "Создатели клипов", дополнительный буст охвата

## Обучение модели

Для построения оптимальной системы рекомендаций и глубокого понимания того, какие товары автор сможет успешно продвигать, мы используем методы машинного обучения. Подход следующий: сначала для каждого товара определяется его категория (подробнее о сегментации расскажем далее). Затем модель рекомендует товар тем авторам, которые исторически показывали высокие результаты в этой категории, а также авторам с похожим профилем и поведенческими метрикам. То есть тем, кто потенциально сможет эффективно продать данный товар.

Подобный процесс можно выполнять вручную, однако человеческий анализ неизбежно ограничится широкими категориями и даст значительно менее точный результат. Система рекомендаций, по сути, является инструментом искусственного интеллекта: она анализирует данные о товарах и авторах, определяет уровень «мэтча» между ними и предсказывает ожидаемое количество просмотров, кликов и заказов. Это позволяет формировать для каждого автора оптимизированную ленту заказов, максимизируя эффективность рекламных размещений.

### Запуск обуения

```python
targets = ['views', 'clicks', 'n_orders']
models = {}

for target in targets:
    print(f'\n--- Обучаем модель для {target} ---')

    y_single = df_grouped[target]

    X_train, X_test, y_train, y_test = train_test_split(
        X, y_single, test_size=0.1, random_state=42
    )

    train_pool = Pool(X_train, y_train, cat_features=cat_features)
    test_pool  = Pool(X_test, y_test,  cat_features=cat_features)

    model = CatBoostRegressor(
        iterations=1800,
        learning_rate=0.03,
        depth=8,
        l2_leaf_reg=4,
        loss_function='RMSE',
        eval_metric='RMSE',
        cat_features=cat_features,
        random_strength=0.7,
        bootstrap_type='Bayesian',
        bagging_temperature=1,
        leaf_estimation_iterations=4,
        verbose=200
    )

    model.fit(train_pool, eval_set=test_pool, use_best_model=True)

    models[target] = model
```
